```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
census = pd.read_csv('nyc_census_tracts.csv')
```

```python
census
```

```python
nyc_total = sum(census.TotalPop.values)
```

```python
tot_wom = census.Women.values.sum()
tot_men = census.Men.values.sum()
```

```python
nyc_women = pd.DataFrame(census.groupby("Borough").Women.sum())
```

```python
nyc_women
```

```python
nyc_women["wom_per"] = [round(wom*100/tot_wom,1) for wom in nyc_women.Women]
```

```python
nyc_women.reset_index(inplace=True)
```

```python
nyc_women.sort_values(by="wom_per",ascending=False,inplace=True)
```

```python
nyc_women["Men"]=list(census.groupby("Borough").Men.sum())
```

```python
nyc_women["men_per"] = [round(men*100/nyc_total,1) for men in nyc_women.Men]
```

```python
nyc_women.sort_values(by="wom_per",ascending=False,inplace=True)
```

```python
plt.bar(nyc_women["Borough"],nyc_women["wom_per"],width=0.6,color=sns.color_palette("OrRd_r"))
plt.xlabel("Borough")
plt.ylabel("Percent of Women",)
plt.title("WOMEN DISTRIBUTION IN NYC",weight="bold",fontsize=12)
plt.savefig("fem_dist.png");
```

```python
nyc_women.sort_values(by="men_per",ascending=False,inplace=True)
```

```python
plt.bar(nyc_women["Borough"],nyc_women["Men"],width=0.6,color=sns.color_palette("OrRd_r"))
plt.xlabel("Borough",weight="bold")
plt.ylabel("Percent of Women")
plt.title("MEN DISTRIBUTION IN NYC",weight="bold",fontsize=12)
plt.savefig("men_dist.png");
```
