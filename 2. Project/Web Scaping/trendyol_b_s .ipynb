```python
from bs4 import BeautifulSoup
import requests
import time, os
```

```python
import pandas as pd
import numpy as np
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

chromedriver = "C:/Users/MERVE/Desktop/chromedriver" # path to the chromedriver executable
os.environ["webdriver.chrome.driver"] = chromedriver
```

```python
query = "laptop"  # trendyolda laptop aratacağım.
trendyol_search = f"https://www.trendyol.com/sr?q={query}&qt={query}&st={query}&os=1"
url_list = [trendyol_search+"&pi="+str(page) for page in range(1,209)] # ilk 208 sayfadaki bilgisayarları alıyorum.
```

```python
computers = []
for url in url_list:
    driver = webdriver.Chrome(chromedriver)
    driver.get(url)
    soup_main = BeautifulSoup(driver.page_source, 'html.parser')
    #her sayfada 24 adet laptop var. 
    links = ["https://www.trendyol.com/"+link.a.get("href") for link in soup_main.find_all("div",{"class":"p-card-chldrn-cntnr card-border"})]
    for link in links:
        driver.get(link)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        brand = soup.find("h1",{"class":"pr-new-br"}).text
        title = soup.find("h1",{"class":"pr-new-br"}).span.text
        try:
            stars = 0
            for star in soup.find("div",{"class":"ratings readonly"}).find_all("div",{"class":"star-w"}):
                stars += int(star.div.findNext().findNext().get("style")[7:10].replace("%","").replace("px",""))            
        except:
            stars = 0
        stars /= 100 
        price = soup.find("span",{"class":"prc-dsc"}).text.split(" ")[0]
        attributes = [(attribute.span.text,attribute.span.findNext().text) for attribute in soup.find("ul",{"class":"detail-attr-container"}).find_all("li",{"class":"detail-attr-item"})]
        attributes = [("Brand",brand), ("Title",title)]+ attributes+[("Star",stars),("Price",price)]
        computers.append(attributes)
driver.quit()
```

## İlk 206 sayfadaki veriyi çektim. Ama bundan sonra Trendyol filtreleme ile arama yapın uyarısı verdiği için error aldım. Yine de üzerinde çalışacabileceğim 4934 tane laptop verisini çekebildim. 

```python
columns= []
for attribute in computers:
    for attribute_name in attribute:
        if attribute_name[0] not in columns:
            columns.append(attribute_name[0])
columns.sort()
```

```python
df = pd.DataFrame(columns = columns)
```

```python
# her bilgisayar için eksik özellikleri NaN ile doldurma
for ix,computer in enumerate(computers):
    existing_attributes = set([attribute[0] for attribute in computer]) # var olan özellikler
    missing_attributes = set(df.columns) - existing_attributes
    for attribute in missing_attributes:
        computer.append((attribute,np.nan))
    computer.sort()
```
